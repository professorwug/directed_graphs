```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Plotting Vector Fields using matplotlib

Source: https://krajit.github.io/sympy/vectorFields/vectorFields.html

```python
# controls the x and y axes of the plot
# linspace(min on axis, max on axis, number of arrows per axis)
x,y = np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20))

#specify direction of vector field
u = -1 # vector field in x direction
v = 1 # vector field in y direction

""" 
quiver 
    plots a 2D field of arrows
    quiver([X, Y], U, V, [C], **kw); 
    X, Y define the arrow locations, U, V define the arrow directions, and C optionally sets the color.
"""
plt.quiver(x,y,u,v)

"""Display all open figures."""
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20))

u = x/np.sqrt(x**2 + y**2)
v = y/np.sqrt(x**2 + y**2)

plt.quiver(x,y,u,v)
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(0,1.5,20),np.linspace(0,1.5,20))

u = x/np.sqrt(x**2 + y**2)
v = y/np.sqrt(x**2 + y**2)

plt.quiver(x,y,u,v)
plt.show()
```
