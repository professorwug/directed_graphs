# Spiral

```python
from directed_graphs.datasets import directed_spiral, plot_directed_2d
X, flow, labels = directed_spiral(num_nodes=1000, num_spirals=2.5, radius=1)
plot_directed_2d(X, flow, labels)
```

```python
X = torch.tensor(X)
flow = torch.tensor(flow)
X = X.float().to(device)
flow = flow.float().to(device)
```

```python
from directed_graphs.diffusion_flow_embedding import DiffusionFlowEmbedder
dfe = DiffusionFlowEmbedder(
  X, flow, t=1, sigma_graph = 1, sigma_embedding= 1, flow_strength_graph=2, flow_strength_embedding=2,learnable_flow_strength=True,
  smoothness=0.01, device=device, weight_of_flow=1, learning_rate = 0.001,
)
dfe.fit(n_steps=5000)
```

```python
sigmas = [0.5, 1, 5, 10]
flow_strengths = [1, 2, 20, 50,100]
affinity_grid_search(X, flow, sigmas, flow_strengths)
```
